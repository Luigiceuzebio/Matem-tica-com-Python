# Teste de convergência

Seja $ X_n $ e $ Y_n $ variáveis aleatórias contínuas independentes e uniformemente distribuidas entre o intervalo $ [-1, 1] $, onde $ n = 1, 2, ...$

Além disso, considere $$ I_n{} = \begin{cases} 1, & \mbox{se } X^2_n + Y^2_n \leq 1  \\ 0, & \mbox{caso contrário.} \end{cases} $$

Nesse caso, tilizando $ n = 10000 $, vamos calcular \begin{equation}P_i = \sum_{i=1}^{n} \frac{I_n}{n}\end{equation}

```python
import numpy as np
import matplotlib.pyplot as plt

def calcula_probabilidade(n=10000):
    X = np.random.uniform(-1, 1, n)  #Função para gerar as variáveis aleatórias entre [-1, 1]
    Y = np.random.uniform(-1, 1, n)  #Função para gerar as variáveis aleatórias entre [-1, 1]
    
    I = (X**2 + Y**2 <= 1).astype(int) 
    
    probabilidade = np.sum(I) / n #Função de somatório da biblioteca Numpy
    return probabilidade, X, Y, I

n = 10000

probabilidade_aproximada, X, Y, I = calcula_probabilidade(n)
print(f'Probabilidade aproximada com n={n}: {probabilidade_aproximada}')

probabilidade_acumulada = np.cumsum(I) / np.arange(1, n+1)  #Dividimos a soma acumulada dos indicadores pelo número de pontos considerados até aquele ponto

plt.figure(figsize=(9, 5))
plt.plot(range(1, n+1), probabilidade_acumulada, label='Probabilidade acumulada')
plt.axhline(y=np.pi/4, color='r', linestyle='--', label='Valor Esperado: π/4')
plt.xlabel('Número de pontos (n)')
plt.ylabel('Probabilidade acumulada')
plt.title('Convergência da probabilidade acumulada para π/4')
plt.legend()
plt.grid(True)
plt.show()
```

É fácil notar que a probabilidade que desejamos calcular tende ao valor $\frac{\pi}{4}$. Isto ocorre pois as variáveis aleatórias contínuas e independentes $ X_n $ e $ Y_n $ estão definidas no intervalo fechado $ [-1, 1] $. Pensando em um plano cartesiano com os eixos $ X_n $ e $ Y_n $, esses pontos estarão em um quadrado de área 4, já que possui lado 2. 
Basicamente, os 10000 pontos gerados estão dentro desse "quadrado". Quando adicionada a restrição $ I_n $, temos pontos que estão dentro ou fora da circuferência de raio 1 com área igual a $\pi$. Então, ao calcularmos a probabilidade $ P_i $ com $ i $ muito grande o resultado tende a razão $\frac{\pi}{4}$. Se observarmos o gráfico acima, quanto mais pontos são plotados, mais precisa é essa aproximação. O contrário também pode ser notado.

```python
pi_aprox = 3.14
resultado = pi_aprox / 4
print(f'O resultado da razão entre pi e 4 eh {resultado}')
print(f'Probabilidade aproximada com n={n}: {probabilidade_aproximada}')
```
